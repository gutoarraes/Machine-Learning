## Convolutional neural network (difference from dense neural network)

*Note: We will use the term convnet and convolutional interchangeably.*
Dense neural network looks at the whole picture to learn patterns. Convolutional looks for local paterns - some features are looked for. In a Dense network the patterns would need to be located in the same place around the image. 

Convolutional neural network returns a map of patterns.
A filter is a pattern that the sequence follows.
each convolutional layers we look for many filters (32, 64 or 128  usually)
stride: the amount you move over X amount of pixels to the next block to be analyzed.
pooling: min) take the minimum value of a section to create another layer that will be half the size of the original map;
         max) take the maximum value of a section to create another layer that will be half the size of the original map;
         average) take the average value of a section to create another layer that will be half the size of the original map;
         Usually we do a 2x2 pooling with a stride of 2. Different kinds of pooling are used for different things. A max pooling will tell you what if a feature is present in that area. A min pooling will tell you if a feature exist.

# Layer 2

# Layer 3
The next set of layers do very similar things but take as input the feature map from the previous layer. They also increase the frequency of filters from 32 to 64. We can do this as our data hrinks in special dimensions as it passed through thte layers, meaning we can afford (computationally) to add more depth.

## Fine Tuning

When we employ the technique defined above we ewill often want to tweak the final layers in our convolutional base to work better for our specific problem. TThis involves not touching or retrtaingin the earlier layers in our convolutional base but only adjusting the final ffew. We do this because the first layers in our base are very good at extracting low level features lile lines and edges, things that are similar for any kind of image. Where the later layers are better at picking up bvery specific features like shapes or even eyes. If we adjust the final layers than we look for only features relevant to our very specific problem

## Data processing

```python
IMG_SIZE = 160 # all images will be resized to 160x160. Better to compress than to make it too big

def format_example(image, label):
  """
  returns an image that is reshaped to IMG_SIZE
  """
  image = tf.cast(image, tf.float32)
  image = (image/127.5) - 1
  image = tf.image.resize(image, (IMG_SIZE, IMG_SIZE))
  return image, label
```

Now we can apply this function to all our images using map

```python
train = raw_train.map(format_example)
validation = raw_validation.map(format_example)
test = raw_test.map(format_example)
```

## Picking a Pretrained Model

The model we are going to use as the convo base for our model is the MobileNet V2 developeed at google. this model is trained on 1.4 milion images and has 1000 different classes.
We want to use this model but only its convolutional base. So when we load in the model we'll specify that we don't want to load the top (classification) layer. We'll tell the model what input shape to expect and to use the predetermined weights from *imagenet* (Google datase)

## Bag of words function

```python
vocab = {} # maps word to integer representing it
word_encoding = 1
def bag_of_words(text):
  global word_encoding

  words = text.lower().split(" ") # create list of all words, we'll assume there is no numbers
  bag = {} # stores all of the encodings and their frequency

  for word in words:
    if word in vocab:
      encoding = vocab[word] # get encoding from vocab
    else:
      vocab[word] = word_encoding
      encoding = word_encoding
      word_encoding += 1

      if encoding in bag:
        bag[encoding] += 1
      else:
        bag[encoding] = 1
  return bag

text = "this is a test to see if this test will work is is test a a"
bag = bag_of_words(text)
print(bag)
print(vocab)
```

## Encoding

Since this text isn't encoded yet well need to do that ourselves. We are going to encode each unique character as a different integer.

```python
vocab = sorted(set(text))
# creating a mapping from unique characters to indices
char2idx = {u:i for i, u in enumerate(vocab)}
idx2char = np.array(vocab) # reverse the list from letter to int, to from int to letter (5:55 of video)

def text_to_int(text):
  return np.array([char2idx[c] for c in text])

text_as_int = text_to_int(text)

```

```python
def int_to_text(ints):
  try:
    ints = ints.numpy()
  except:
    pass
  return ''.join(idx2char[ints])

print(int_to_text(text_as_int[:13]))
```

## Building the model

Now it's time to build the model. We will use an embedding layer a LSTM and one dense layer that contains a node in our training data. The dense Layer will give us a probablity distribution over all nodes.

```python
def build_model(vocab_size, embedding_dim, rnn_units, batch_size):
  model = tf.keras.Sequential([
          tf.keras.layers.Embedding(vocab_size, embedding_dim,
                                    batch_input_shape=[batch_size, None]),
          tf.keras.layers.LSTM(rnn_units,
                               return_sequences=True,
                               stateful=True,
                               recurrent_initializer='glorot_uniform'),
                               tf.keras.layers.Dense(vocab_size)
  ])
  return model

model = build_model(VOCAB_SIZE, EMBEDDING_DIM, RNN_UNITS, BATCH_SIZE)
model.summary()
```

## Learning the Q-Table

Now how do we create this table and find those values. Well this is where we will discuss how the Q-learning algorithm updates the values in our Q-Table.
I'll start by noting that our Q-Table starts with all 0 values. This is because the agent has yet to learn anything about the environmnet. Our agent learns by exploring the environment and oserving the outcome/reward from each action it takes in a given state. But how does it know what action to take in each state? There are two ways that our agent can decide on which action to take. 
  1. Randomly picking a valid action
  2. Using the current Q-Table to find the best action

Near the beginning of our agents learning it will mostly take random actions in order to explore the environment many different states. As it starts to explore more of the environment it will start to gradually rely more on it's learned values (Q-Table) to take actions. This mean s that as our agent explores more of the enviromnet it will develop a better understanding a start to take "correct" or better actions more often. It's important thatt the agent has a good balance of taking random actions and using learned values to ensure it does get trapped in a local maxima.
After each new action our agent will record the new state (if any) that it has entered and the reard that it received from taking that action. These values will be used to update the Q-Table. The agent will stop taking new actions after a certain time limit is reached or it has achieved the goal or reached the end of the environmnet.

## Updating Q-Values:
 
 Q[state,action] = Q[state,action] + alfa * (reward + gamma*max(Q[newstate,:] - Q[state,action]) 

 # Learning Rate | alfa
 The learning rate is a numeric constant that defines how much change is permitted on each QTable update. A high learning rate means that each update will introduce a large change to the current state-action value. A small learning rate means that each update has a more subtle change. Modifying the learning rate will change how the agent explorees the environment nd how quickly it dfetermines the final values in the QTable.

 # Discount factor | Gamma
 Discount factor is used to balance how much focus is put on the current and future reward. A high discount factor means that future rewards will be considered more heavily. 
 To perform updates on this table we will let the agent explore the enviromnet for a certain period of time and use each of its actions to make an update. Slowly we shoulde start to notice the agent learning and choosing better options.

## Putting it all together

Now that we know how to do some basic things we can combine these together to create our Q-Learning Algorithm

```python
import gym
import numpy as np
import time

env = gym.make("FrozenLake-v0")
STATES = env.observation_space.n
ACTIONS = env.action_space.n

Q = np.zeros((STATES, ACTIONS)) # create a matrix with all 0 values

EPISODES = 10000 # How many times to run the environment from the begginning
MAX_STEPS = 100 # max number of steps allowed for each run of environment

LEARNING_RATE = 0.81
GAMMA = 0.96

RENDER = False # If you want to see training set to True
epsilon = 0.9
```

```python
rewards = []
for episode in range(EPISODES):

  state = env.reset()
  for _ in range(MAX_STEPS):

    if RENDER:
      env.render()

    if np.random.uniform(0, 1) < epsilon:
      action = env.action_space.sample()
    else:
      action = np.argmax(Q[state,:])

    next_state, reward, done, _ = env.step(action)

    Q[state, action] = Q[state, action] + LEARNING_RATE * (reward + GAMMA * np.max(Q[next_state,:]) - Q[state, action])

    state = next_state
    if done:
      rewards.append(reward)
      epsilon -= 0.001
      break # reached goal

print(Q)
print(f"average reward: {sum(rewards)/len(rewards)}:")
# Now you can see our Q values!
```
